```python
##Subroutine calculates how many samples are associated to each node
def consolida_nodes(node1):
    nodes = pd.DataFrame (columns = ['number','par1','par2','par3','par4','par5','par6','par7','par8','par9','par10','label1x','label2x','label3x','label4x','label5x','label6x','label7x','label8x','label9x','label10x','label11x','label12x','total','total_per'])
    ##print(nodes)
    ## Here we convert the Kmeans format into matrix adding the necessary columns
    len2 = len(node1)
    #print(len2)
    #print(node1)
    #print(node1[9][7])
    for ix in range(len2):
        #print(ix)
        new_row = {'number': ix, 'par1':node1[ix][0], 'par2':node1[ix][1], 'par3':node1[ix][2],'par4':node1[ix][3],'par5':node1[ix][4], 'par6':node1[ix][5],'par7':node1[ix][6],'par8':node1[ix][7],'par9':node1[ix][8],'par10':node1[ix][9], 'label1x':0,'label2x':0,'label3x':0,'label4x':0,'label5x':0,'label6x':0,'label7x':0,'label8x':0,'label9x':0,'label10x':0,'label11x':0,'label12x':0,'total':0,'total_per':0}
        #append row to the dataframe
        nodes = nodes.append(new_row, ignore_index=True)

    ## clusters is the nodes matrix where we are going to place positive and negative for reach and sell
    len3 = len(train)
    for i in range(len3):  ## All trainning data records
  
      for ix in range(len2):  ## all nodes 
    
        if train['node'][i]==nodes['number'][ix]: ## if the node is the one associates with the item
      
            if train['label1'][i] == 0:
                nodes['label1x'][ix] = nodes['label1x'][ix]+1 ## positive for item
      
            if train['label1'][i] == 1:
                nodes['label2x'][ix] = nodes['label2x'][ix]+1 ## positive for item
            if train['label1'][i] == 2:
                nodes['label3x'][ix] = nodes['label3x'][ix]+1 ## positive for item
            if train['label1'][i] == 3:
                nodes['label4x'][ix] = nodes['label4x'][ix]+1 ## positive for item
            if train['label1'][i] == 4:
                nodes['label5x'][ix] = nodes['label5x'][ix]+1 ## positive for item
            if train['label1'][i] == 5:
                nodes['label6x'][ix] = nodes['label6x'][ix]+1 ## positive for item
            if train['label1'][i] == 6:
                nodes['label7x'][ix] = nodes['label7x'][ix]+1 ## positive for item
            if train['label1'][i] == 7:
                nodes['label8x'][ix] = nodes['label8x'][ix]+1 ## positive for item
            if train['label1'][i] == 8:
                nodes['label9x'][ix] = nodes['label9x'][ix]+1 ## positive for item
            if train['label1'][i] == 9:
                nodes['label10x'][ix] = nodes['label10x'][ix]+1 ## positive for item
            if train['label1'][i] == 10:
                nodes['label11x'][ix] = nodes['label11x'][ix]+1 ## positive for item
            if train['label1'][i] == 11:
                nodes['label12x'][ix] = nodes['label12x'][ix]+1 ## positive for item
      
            nodes['total'][ix]= nodes['total'][ix]+1
      

        ##
    nodes['label1x']=nodes['label1x']/nodes['total']
    nodes['label2x']=nodes['label2x']/nodes['total']
    nodes['label3x']=nodes['label3x']/nodes['total']
    nodes['label4x']=nodes['label4x']/nodes['total']
    nodes['label5x']=nodes['label5x']/nodes['total']
    nodes['label6x']=nodes['label6x']/nodes['total']
    nodes['label7x']=nodes['label7x']/nodes['total']
    nodes['label8x']=nodes['label8x']/nodes['total']
    nodes['label9x']=nodes['label9x']/nodes['total']
    nodes['label10x']=nodes['label10x']/nodes['total']
    nodes['label11x']=nodes['label11x']/nodes['total']
    nodes['label12x']=nodes['label12x']/nodes['total']
    nodes['total_per']=nodes['total']/size
    return(nodes)
#nodes.to_csv('/Curso-ML/Assignment-3/nodes.csv')
```

```python
##Subroutine that verifies if the clustering matches the threshold (Minimum % of samples per node)
def threshold(threshold1):
    requirement = 0
    for v in range(len(nodes)):
        print(nodes['total'][v],nodes['total_per'][v])
        if nodes['total_per'][v]<threshold1:
            requirement=1
    return requirement
```

```python
##Subroutine that verifies if the clustering matches the threshold (Minimum % of samples per node)
def thresholdx1(threshold2,threshold3):
   
    requirement = 1
    count_melhores=0
    for v in range(len(nodes)):
        #print(nodes['total'][v],nodes['total_per'][v])
        if nodes['melhores'][v]>threshold2:
            count_melhores=count_melhores+1
    #print('% melhores: ',count_melhores,len(nodes),count_melhores/len(nodes),threshold3)
    percent_melhores=count_melhores/len(nodes) ## % of the nodes reach the criteria
        
    if percent_melhores>threshold3: ## if percentage of melhores reach the definied limit
        print('% melhores: ',count_melhores,len(nodes),count_melhores/len(nodes),threshold3)
        requirement=0
    return requirement,percent_melhores
```

```python
# Select the three best labels and sum it into a variable
def melhores():
    nodes['bestlabelx1']=''
    nodes['bestlabelx2']=''
    nodes['bestlabelx3']=''
    nodes['bestlabel1'] =0
    nodes['bestlabel2'] =0
    nodes['bestlabel3'] =0
    nodes['melhores']   =0.00
    
    var1=[]
    var1=['label1x','label2x','label3x','label4x','label5x','label6x','label7x','label8x','label9x','label10x','label11x','label12x']
    len3 = len(nodes)
    for i1 in range(len3):
        maior1 = 0
   
   
        labelxxx1=""
        labelxxx2=""
        labelxxx3=""
        for i in range(12):
            var2=nodes[var1[i]][i1]
            if maior1<var2:
                maior1=var2
                #print(var2)
                labelxxx1 = var1[i]
        maior2 = 0
        for i in range(12):
            var2=nodes[var1[i]][i1]
            if maior2<var2 and var2<maior1:
                maior2=var2
                #print(var2)
                labelxxx2 = var1[i]
        maior3 = 0
        for i in range(12):
            var2=nodes[var1[i]][i1]
            if maior3<var2 and var2<maior2:
                maior3=var2
                #print(var2)
                labelxxx3 = var1[i]
            
        #print(maior1,maior2,maior3)
        nodes['bestlabelx1'][i1]=labelxxx1
        labely1 = labelxxx1[5:]
        labely1 = labely1[:-1]
        nodes['bestlabel1'][i1]=int(labely1)-1
        #print(labelxxx1)
    
        nodes['bestlabelx2'][i1]=labelxxx2
        labely2 = labelxxx2[5:]
        labely2 = labely2[:-1]
        nodes['bestlabel2'][i1]=int(labely2)-1
        #print(labelxxx2)
    
        nodes['bestlabelx3'][i1]=labelxxx3
        labely3 = labelxxx3[5:]
        labely3 = labely3[:-1]
        nodes['bestlabel3'][i1]=int(labely3)-1
        #print(labelxxx3)
        maioresx = maior1+maior2+maior3
        nodes['melhores'][i1]=maioresx
        
```

```python
### Clusterizing the samples
train['node']=0.0
train['node1']=''
size = len(train) ## Number of samples 
threshold1 = 0.05 ### MInimum percentage per node
threshold2 = .40 ##Percentage of type of the labels
threshold3 = .50 ##Percentage of the nodes archieving the label
maior_percentage=0
melhor_node = 0
X = np.array(train.drop(['label1','node','node1'], 1).astype(float))
y = np.array(train['label1'])
train.info()
### Clusterization
nodes_num=50
while nodes_num>12:
    print(nodes_num)
    
    kmeans = KMeans(algorithm='auto', copy_x=True, init='k-means++', max_iter=300,n_clusters=nodes_num, n_init=10, n_jobs=1, 
               precompute_distances='auto',random_state=None, tol=0.0001, verbose=0) # Number of nodes
    kmeans.fit(X)
    


    
    ## Generates a vector indicating the node associated to each row
    node2 = kmeans.labels_
    atribui(node2)
    ## Check how many samples are associated to each node
    node1 = kmeans.cluster_centers_
    nodes=consolida_nodes(node1)
    #identifies percentage per label
    melhores()
    ## Check the percentage of samples per node is bigger than a threshold
    ##requirement2 = threshold(threshold1)
    requirement2, per_melhores = thresholdx1(threshold2,threshold3)
    ## if threshold not achieved reduce the number of nodes and run again
  
    if requirement2 == 0:
        print(per_melhores)
        if per_melhores>maior_percentage:
            maior_percentage=per_melhores
            melhor_node =nodes_num
            print('Best alternatives until now: ',melhor_node,'Best % concentration: ',maior_percentage)
        #break
    nodes_num = nodes_num-1
##node1 <- fitted(valor, method = c("centers"))
print(maior_percentage)
print(melhor_node)
```
