# Identifying to which node the test item is associated

```python
len1 = len(df2)
contador0 = 0 ## paginacao
contador1 = 0 
page=100
import datetime
import time
ts = time.time()
st = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
print(st)
while (contador1<=len1-1): ## Loop items test set
    
    #print(contador1) 
    
    
    
    
    if (contador0 == page):
      page = page+100
      print(contador1)
      ts = time.time()
      st = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
      print(st)
    
    contador0=contador0+1
    
 
    tpar1   = df2['par1'][contador1]
    tpar2   = df2['par2'][contador1]
    tpar3   = df2['par3'][contador1]
    tpar4   = df2['par4'][contador1]
    tpar5   = df2['par5'][contador1]
    tpar6   = df2['par6'][contador1]
    tpar7   = df2['par7'][contador1]
    tpar8   = df2['par8'][contador1]
    tpar9   = df2['par9'][contador1]
    tpar10  = df2['par10'][contador1]
    
    
    
    
    menor = 10000000
    len2 = len(df1)
    ##print(len2)
    contador2 = 0
    ##print("Inicio ciclo clusters")
    while (contador2 <= (len2-1)):
        ##print(contador2)

        ttpar1   = df1['par1'][contador2]
        ttpar2   = df1['par2'][contador2]
        ttpar3   = df1['par3'][contador2]
        ttpar4   = df1['par4'][contador2]
        ttpar5   = df1['par5'][contador2]
        ttpar6   = df1['par6'][contador2]
        ttpar7   = df1['par7'][contador2]
        ttpar8   = df1['par8'][contador2]
        ttpar9   = df1['par9'][contador2]
        ttpar10  = df1['par10'][contador2]
      
        ## Subtraction section 
      
        tttpar1  = ttpar1 - tpar1
        tttpar2  = ttpar2 - tpar2
        tttpar3  = ttpar3 - tpar3
        tttpar4  = ttpar4 - tpar4
        tttpar5  = ttpar5 - tpar5
        tttpar6  = ttpar6 - tpar6
        tttpar7  = ttpar7 - tpar7
        tttpar8  = ttpar8 - tpar8
        tttpar9  = ttpar9 - tpar9
        tttpar10 = ttpar10 - tpar10
    
        quadrado = ((tttpar1**2)+(tttpar2**2)+(tttpar3**2)+(tttpar4**2)+(tttpar5**2)+(tttpar6**2)+(tttpar7**2)+(tttpar8**2)+(tttpar9**2)+(tttpar10**2))
      
        d = (quadrado**1/2)
      
        if (d<menor and d!=0): ## selection of the best cluster (small distance)
            ##print(d)
            registro1 = df1['Id'][contador2] 
            registro  = contador2
            menor = d
      
      
        contador2 = contador2 + 1
        ## end loop contador 2
    
    df2['label1x'][contador1] = df1['label1x'][registro]##
    df2['label2x'][contador1] = df1['label2x'][registro]##
    df2['label3x'][contador1] = df1['label3x'][registro]##
    df2['label4x'][contador1] = df1['label4x'][registro]##
    df2['label5x'][contador1] = df1['label5x'][registro]##
    df2['label6x'][contador1] = df1['label6x'][registro]##
    df2['label7x'][contador1] = df1['label7x'][registro]##
    df2['label8x'][contador1] = df1['label8x'][registro]##
    df2['label9x'][contador1] = df1['label9x'][registro]##
    df2['label10x'][contador1] = df1['label10x'][registro]##
    df2['label11x'][contador1] = df1['label11x'][registro]##
    df2['label12x'][contador1] = df1['label12x'][registro]##
    df2['node'][contador1] = registro
    df2['bestlabelx1'][contador1]=df1['bestlabelx1'][registro]
    df2['bestlabel1'][contador1] =df1['bestlabel1'][registro]
    df2['bestlabelx2'][contador1]=df1['bestlabelx2'][registro]
    df2['bestlabel2'][contador1] =df1['bestlabel2'][registro]
    df2['bestlabelx3'][contador1]=df1['bestlabelx3'][registro]
    df2['bestlabel3'][contador1] =df1['bestlabel3'][registro]
    contador1 = contador1+1
    ##print(contador1)
    
## end loop items contador1
```
